```python
import nltk
# nltk.download('punkt')
# from nltk.tokenize import word_tokenize

nltk.download('stopwords' )
from nltk.corpus import stopwords

# nltk.download( 'wordnet' )
# from nltk import pos_tag
# from nltk.corpus import wordnet

import pandas as pd
import io
import requests
```

```python
df = pd.read_csv('3.2_translated_data_g20.csv')
df.head()
```

```python
# extracting the stopwords from nltk library
sw = stopwords.words('english')
# sw.extend(["yg", "dg", "rt", "dgn", "ny", "d", 'klo', 
#                        'kalo', 'amp', 'biar', 'bikin', 'bilang', 
#                        'gak', 'ga', 'krn', 'nya', 'nih', 'sih', 
#                        'si', 'tau', 'tdk', 'tuh', 'utk', 'ya', 
#                        'jd', 'jgn', 'sdh', 'aja', 'n', 't', 
#                        'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt',
#                        '&amp', 'yah'])
def stopwordsext(text):
    '''a function for removing the stopword'''
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)

df['tweet_en'] = df['tweet_en'].apply(stopwordsext)
df['tweet_en'].head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud


allWords  = ' '.join( [str (twts) for twts in df['tweet_en']])
wordCloud = WordCloud(width = 1000, height=800, random_state=21, max_font_size=400).generate(allWords)

plt.imshow(wordCloud, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```
